```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 8] # Dette er for  f en strre figur

hastighet_100 = 100/3.6  # gjr 100 km/t om til m/s
t_0_100 = 4.4      # tiden blien skulle bruke p 0-100 km/t
tidsvektor = np.arange(0, 5.1, 0.5)      # 0 til 5 sek med halv sek opplsning

akselerasjon = hastighet_100 / t_0_100    # beregner akselerasjon
strekning = 0.5 * akselerasjon * tidsvektor**2    # beregner strekningen i meter
# Strekning blir en array med like mange elementer som tidsvektor
fart_kmt = akselerasjon * tidsvektor * 3.6 # beregner hastighet i km/t

plt.plot(tidsvektor, strekning, label='Strekning (m)')
plt.plot(tidsvektor, fart_kmt, label='Hastighet (Km/t)')

plt.grid()      # legger p rutenett
plt.axhline(c='k', lw=2)     # uthever x-aksen
plt.axvline(c='k', lw=2)     # uthever y-aksen
plt.title(" Fart og strekning kjrt ved konstant akselerasjon")
plt.xlabel("sekunder fra start")     # tekst p x-aksen
plt.ylabel("Strekning i meter og fart i km/t")     # tekst p x-aksen
plt.legend()     # viser figurforklaring kurvetyper (farge) med label 
plt.show()       # viser figuren
```

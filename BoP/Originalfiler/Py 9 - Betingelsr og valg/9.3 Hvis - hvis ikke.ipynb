## Hvis - hvis ikke

En av de viktigste brukene av betingelser i programmering er å bestemme om en del av (eller hvilken del av) koden som skal kjøres.  

Hvis betingelsen er oppfyllt så kjøres koden. Hvis ikke hoppes den over.  
Alternativt kjøres en del av koden hvis betingelsen er oppfyllt og en annen del hvis den ikke er det.

### if og if-else

Python har **if** - **if - else** setningen (konstruksjonen) til denne bruk. Vi finner den i så godt som alle programmeringsspråk med velig små forskjeller.

Syntaks for if-setningen:
```Python
if logisk betingelse:  # husk kolon, Dersom betingelsen er sann, så…
    kommando 1 # … utfør "kommando 1".
else: # husk kolon
    kommando 2 # … utfør "kommando 2".
```
+ Det må være kolon etter betingelsen og etter else
+ Kommando 1 utføres bare hvis betingelsen er **sann**
+ Kommando 2 utføres bare hvis betingelsen er **usann**
+ Kommando 1 og 2 er helt vanlig(e) programsetninger
+ **else** delen er bare med dersom vi trenger den
+ Innrykkene er viktig. Det er de som styrer hva som som hører til if/else og resten av koden

Eksempel:

````python
alder = int(input('Din alder?: ') )
if alder < 18:
    print('Du er ikke myndig!') 
else:
    print('Du er myndig!') 
````

### Flytskjema for if og if-else

### Programflyt for if og if-else

- Når programmet kommer til en if, evalueres betingelsen
- Dersom betingelsen er sann
    - Gjøres programmsetningene som er rykket inn  
        - mellom if og else, dersom det er en else
        - etter if dersom det ikke er en else
- Dersom betingelsen er usann
    - Gjøres programmsetningene 
        - etter else, dersom det er en else
    - Avsluttes if-setningen dersom det ikke er en else
- Uansett fortsetter programmet etter de innrykkede setningene som hører til denne if  


#### Ekempler på bruk

Mange bruksområder for if
- Velge alternativer
- Unngå uheldige situasjoner
    - Deling på null eller andre situasjoner der vi ikke får gyldige svar  
    
Her bruker vi det for å unngå å få komplekst tall til svar

````python
# Dette skriptet regner ut kvadratroten et tall som leses inn,
# så lenge tallet er positivt. Dersom tallet er negativt
# skrives det ut en advarsel. Slik unngås komplekst resultat

import numpy as np

tall = int( input('Skriv inn et tall:') )
if tall >= 0:
    rot = np.sqrt(tall)
    print('Kvadratroten av', tall,'er', round(rot,4) )
else:
    print('Tallet er negativt. Det vil gi en kompleks rot.')
````

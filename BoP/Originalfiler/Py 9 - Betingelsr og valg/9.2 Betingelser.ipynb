## Betingelser

- Betingelser er enten oppfyllt (sanne) eller ikke (usanne)  
    - SANN har verdien True  
    - USANN har verdien False  

Dette er altså boolske tilstander/verdier som kan tilordnes variable som da får datatypen bool.

Nå vi samenligner to verdier  stiller vi et konkret spørsmål  
For eksempel:  
- Er A større enn B?  

Det vi i praksis gjør er å komme med en påstand (A (er)større enn B).  
Påstanden er da enten sann eller usann. Det kan være greit og først skrive betingelsen med ord og så gjøre den om til programkode. I koden brukes sammenlikningsoperatorer.

Python har disse sammenlikningsoperatorer:  



Vi kan prøve dette ved å skrive betingelser direkte i celler

````python
2 > 3  # Påstand 2 er større enn 3
````

Selvfølgelig er ikke 2 større enn 3 så vi får usann som resultat. Betingelsen er ikke oppfyllt.

````python
4 == 4  # Påstand 4 er lik 4
````

Her er betingelsen oppfyllt slik vi kan se.  

Det gir egentlig ikke så mye mening i å sammenligne faste tall. Når vi sammenligner variabler som får verdier under kjøring av programmet blir det mer fornuftig.

````python
import random as rnd
rnd.seed()

# a og b settes til tilfeldig tall 0-10
a = rnd.randint(0, 10)
b = rnd.randint(0, 10)

bet = a < b # Påstår at (spør om) a er mindre enn b. Tilordner svaret til bet

print('a =', a)
print('b =', b)

print('Påstanden a < b er', bet)
print(type(bet))
````

Her vil vi få forskjellig resultat når vi kjører koden flere ganger.

### Operatorene

- Sammenligningsoperatorer 
    - De seks første i tabellen  
    - Sammenligner to (og kun to) verdier  
- Logiske operatorer  
    - De tre siste i tabellen  
    - Kan kun operere på SANN/USANN (betingelser)  

- "Logisk og" – betingelse (and)  
    - Gir SANN bare hvis både uttrykkene på høyre og venstre side er SANN  
- "Logisk eller" – betingelse (or)  
    - Gir SANN hvis uttrykket på høyre eller venstre side er SANN  
- "Logisk ikke" – betingelse (not)  
    - Opererer på kun ett uttrykk/betingelse  
    - Gir det motsatte  som resultat

#### Mellom

Python har en mulighet når det gjelder sammenligning som man ikke finner i andre programmeringsspråk.  
Man kan sammenligne tre vedier.
```Python
a < b < c  # Her sjekker man om a er mindre enn b og b er mindre enn c i en operasjon
```  

I Py.9 Øving er ikke denne tatt med som et gyldig svaralternativ.

### Sammensatte betingelser

Sammenligningsoperatorerene brukes i "enkle" sammenligninger av to verdier.  Men noen ganger er betingelsen for det vi skal gjøre mer kompleks.  Da kan vi sette samen flere slike enkle betingelser/sammenligninger med **og**/**eller**. Vi kan også snu et resultat med **ikke**.  

Eksempel:  
A må være mindre enn ti og samtidig må  B være forskjellig fra (ikke lik)  5.  
```Python
A < 10 and B != 5
```


Siden de logiske operatorene har høyere prioritet enn sammenligningsoperatorene trenger vi ikke parentes her.

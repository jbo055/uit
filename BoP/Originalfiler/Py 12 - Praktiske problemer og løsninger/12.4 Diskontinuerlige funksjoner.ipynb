## Diskontinuerlige funksjoner og funksjoner med delt forskrift

Dette er også funksjoner der vi har et «problem» underveis som vi må ta hensyn til.

#### Diskontinuerlige funksjoner

I den forrige notatboken så vi på deling på null der vi kunne oppfatte funksjonen som kontinuerlig. Hvis funksjonen er diskontinuerlig må vi gjøre noe tilsvarende, men vi må passe på at vi ikke trekker en linje over diskontinuiteten. Det gjør vi ved å ha et eget kall til plot for hver del.  

Det lar seg gjøre ved å bruke en forløkke, men er enda enklere om man benytter en metode som gjør at man kan jobbe med en del av arrayen i gangen basert på en betingelse. ```x[x < 0]``` betyr den delen av arrayen x der x < 0. Vi må bruke det samme når vi skal angi den delen av y som skal få resultatet. Altså ```y[x < 0]```. Vi å også ha separate kall til plot() for å unngå at det tegnes en strek mellom endepunktene.

En typisk diskontinuerlig funksjon er $  f(x)= \frac{1}{x} $ . Å plotte denne blir da:

````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 8] # Dette er for å få en større figur

plottpunkter = 200

x = np.linspace(-1.5, 1.5, plottpunkter)
y = np.zeros(plottpunkter) # y blir like stor som x og fyllt med 0
y[x<0] = 1/x[x<0] # [x<0] betyr den delen v arrayen der der x < 0
plt.plot (x[x<0], y[x<0])
# Vi hopper bare over x der vi har diskontinuitet
y[x>0] = 1/x[x>0]# [x>0] betyr den delen v arrayen der der x > 0
plt.plot (x[x>0], y[x>0])
plt.grid()
plt.title('Diskontinuerlig')
plt.xlabel('Som forventet')
plt.ylabel('1/x')
plt.show()
````

#### Funksjoner med delt forskrift¶

Dette er situasjoner der det gjelder forskjellige funksjonsbeskrivelser for forkjellig deler. Om funksjonen er kontinuerlig eller ikke avhenger av om funksjonsverdien er den samme for begge funksjonene i overgangspunktet.  

Som eksempel her bruker vi to polynomer. Det ene gjelder for x <= 2 det andre for x >= 2.  

$ \large\ f(x)={\left\{\begin{matrix}
 2x^{2}-x-3,  x\leq 2\\
\frac{1}{2}x^{2}+2x+3, x\geq 2
\end{matrix}\right.} $  

Det er selvsagt mulig å håndtere dette som å plotte to forskjellige kurver i samme figur med egne array for x- og y-verdier. Men det går også å bruke bare en array for x-verdiene og en for y-verdiene. Vi må da dele opp utregningen. Vi bruker samme teknikk som i forrige eksempel.

````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 8] # Dette er for å få en større figur

plottpunkter = 200

x = np.linspace(-2, 6, plottpunkter)
y = np.zeros(plottpunkter) # y blir like stor som x og fyllt med 0
y[x<=2] = 2*x[x<=2]**2 - 1*x[x<=2] - 3 # [x<=2] betyr den delen v arrayen der der x < 2
plt.plot (x[x<=2], y[x<=2])

y[x>=2] = -1/2*x[x>=2]**2 + 2*x[x>=2] + 3 # [x>=2] betyr den delen v arrayen der der x > 2
plt.plot (x[x>=2], y[x>=2])
plt.grid()
plt.title('Delt forskrift')
plt.xlabel('Som forventet')
plt.ylabel('funksjonsverdier')
plt.show()
````

Vi ser her at det ble et tydelig gap mellom de to funksjonene for x = 2. Dersom vi justerer konstantleddet i de to funksjonene kan vi få de til å møtees i samme punkt.  

$ \large\ f(x)={\left\{\begin{matrix}
 2x^{2}-x-2,  x\leq 2\\
\frac{1}{2}x^{2}+2x+2, x\geq 2
\end{matrix}\right.} $


````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 8] # Dette er for å få en større figur

plottpunkter = 200

x = np.linspace(-2, 6, plottpunkter)
y = np.zeros(plottpunkter) # y blir like stor som x og fyllt med 0
y[x<=2] = 2*x[x<=2]**2 - 1*x[x<=2] - 2 # [x<=2] betyr den delen v arrayen der der x < 2
plt.plot (x[x<=2], y[x<=2])

y[x>=2] = -1/2*x[x>=2]**2 + 2*x[x>=2] + 2 # [x>=2] betyr den delen v arrayen der der x > 2
plt.plot (x[x>=2], y[x>=2])
#plt.plot (x, y)
plt.grid()
plt.title('Delt forskrift')
plt.xlabel('Som forventet')
plt.ylabel('funksjonsverdier')
plt.show()
````

Det kan se ut som om de ikke helt møtes, men de har samme funksjonsverdi for x = 2. Det er bare plotteteknisk at det blir slik.  

Siden de nå møtes i samme punkt kan vi klare oss med ett kall til plot(). Da blir hele kurven tegnet med same farge.

````python
import numpy as np
import matplotlib.pyplot as plt

plottpunkter = 200

x = np.linspace(-2, 6, plottpunkter)
y = np.zeros(plottpunkter) # y blir like stor som x og fyllt med 0
y[x<=2] = 2*x[x<=2]**2 - 1*x[x<=2] - 2 # [x<=2] betyr den delen v arrayen der der x < 2

y[x>=2] = -1/2*x[x>=2]**2 + 2*x[x>=2] + 2 # [x>=2] betyr den delen v arrayen der der x > 2
plt.plot (x, y)
plt.grid()
plt.title('Delt forskrift')
plt.xlabel('Som forventet')
plt.ylabel('funksjonsverdier')
plt.show()
````

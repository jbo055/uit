## Gjensyn med kvadratrotutfordringen

### Hva har vi og hva kan vi gjøre

####  Konkret eksempel

Hvordan regner dere ut kvadratroten av et tall?
    Selvfølgelig bruker dere kvadratrotknappen på kalkulatoren.
- Noen ganger er ikke de enkle mulighetene der
    - Skal se på et konkret eksempel der dette var et stort problem
    - Utgangspunkt: spørsmål fra hovedoppgavegruppe elektronikk

####  Konkret utfordring

- Utfordringen kom fra en Bacheloroppgavegruppe på elektronikk
    - Skulle lage et posisjoneringssystem for dykkere fra en båt
        - Skjerm som viser posisjon til dykker horisontalt, med visning av dybde
    - Dykker har transponder som sender dybde
    - Mottaker på båten måler da avstand (diagonalt) og retning
- Skulle bruke en programmerbar mikrokontroller (singleboardcomputer)
    - Trengte å bruke Pytagoras til å regne ut en katet når den andre katet og hypotenus var kjent (målt)
    - Programmeringsspråket i det aktuelle tilfelle hadde ikke kvadratrotfunksjon eller flyttall
- Hvordan regner vi ut kvadratroten av et tall
    - Når vi ikke bare kan trykke på en knapp
    - eller kalle en ferdig funksjon

####  Håndtering av oppgaven

- Beregne horisontal avstand er helt klart en mindre del av hele oppgaven
    - Oppsplitting av oppgaven i mindre biter
    - Gjentas til man har løsbare deloppgaver
- To viktige spørsmål
    - Hva vet vi?
    - Hva trenger vi?
- Løsning
    - Hvordan bruker vi det vi vet til å finne det vi trenger

#### Spørsmål og svar

- Hva vet vi?
    - Dybde
    - Avstand
- Hva trenger vi?
    - Horisontal avstand
- Løsning
    - Pytagoras
    - katet = dybde
    - hypotenus = avstand
- Hva mer vet vi?
    - horisontal^2 = avstand^2 – dybde^2
    - Avstand >= dybde
- Er kvadrering et problem?
- Utfordringen vi står igjen med
    - Finne kvadratrot

#### Kvadratroten

- Hva vet vi?
    - $ \sqrt[]{1} = 1 $
    - $ \sqrt[]{x>1} => 1 < svar < x $
    - $ \sqrt[]{x>4} => 1 < svar < x/2 $

#### Løsning

- Kan vi
    - gjette på et tall?
        - Vilkårlig tall (random)
    - telle?
- Bedre metoder?
- Alle disse har noen fellestrekk
    - Framgangsmåter som har betydning for programmeringen
    - Sjekke om vi har funnet svaret
    - Hvis svaret ikke er funnet — prøv på nytt (gjentakelser)

####  Må bestemme når svaret er funnet

- Når er vi nært nok svaret?
- I første omgang bruker vi bare hele tall
- For kvadrattall får vi eksakt svar
- For andre tall
    - Det nærmeste hele tallet som er mindre/større

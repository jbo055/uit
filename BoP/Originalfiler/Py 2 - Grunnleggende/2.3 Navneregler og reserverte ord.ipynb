## Navneregler, reserverte ord, anbefalinger og kommentarer

Navn gir vi til variabler, funksjoner og andre ting vi lager.<br>
Python, som alle andre programmeringsspråk, har regler for hva som er lov og ikke lov når det gjelder navn vi gir.
### Følgende regler gjelder:
- Et navn kan ikke begynne med tall
- Et navn kan ikke inneholde mellomrom
- Et navn kan bare inneholde:
    - bokstaver
        - store og små
        - bare a til z (ikke æøå)
    - tall
    - understrekingstegnet _

Python gjør forskjell på store og små bokstaver slik at Tall, tall og TALL er tre forskjellige navn.

### Anbefalinger som man (absolutt) bør følge
I tilleg til de reserverte ordene har vi navnet på funksjoner, som vi kommer til å bruke en del av. De er ikke reservert på samme måte, men har sin spesielle bruk og bør unngås som navn på linje med de reserverte.  

De fleste reserverte ord (og navn på funksjoner) skrives med bare små bokstaver. For å unngå forvirring bør man heller ikke bruke dem som navn med stor forbokstav eller bare store bokstaver. Navnene i lista som er skrevet med storforbokstav er forhåndsdefinerte verdier. Det samme gjelder disse. 

**Generelle anbefalinger**  
I eksemplene vi har sett tidligere har vi brukt variabelnavn som *a*, *b* og *c*. Det er greit når man holder på med ren matematikk der innholdet i variablene ikke har noen bestemt betydning. **Men**  generelt gjelder at man skal gi beskrivende navn til variabler, funksjoner med mer. Dette vil gjøre koden klarere og lettere å forstå.<br>

Hvis vi ikke bruker etablerte navn fra for eksempel fysikk og matematikk, slik det er gjort i eksemplet under, så bør man begynne et variabelnavn med liten bokstav. trenger man å sette sammen variabelnavnet av flere ord/deler kan man bruke understrekingstegnet **_** . Eller ha stor forbokstav på hver del unntatt det første. For eksempel: fornavnOgEtternavn

**Eksempel:** 
Førsteårsstudenten Anna har har en kokeplate på hybelen av den gode gamle sorten. Men hun synes den er dårligere enn forventet. Den har en oppgitt effekt på **1500W** ved en nominell spenning på **240V**. Hun måler spenningen til bare **226V** og vil regne ut hvor stor effekt hun da får.<br>

Fra fysikken kjenner vi effektformelen og Ohms lov med **P = U * I** og **U = R * I** hvor vi har **P** for effekt, **U** for spenning, **I** for strøm og **R** for resistans (motstand). Dette kan vi bruke for å beregne effekten ved den aktuelle spenningen.<br>

Vi vil gi variablene enda mer beskrivende navn enn de korte fra fysikken. I tillegg introduserer vi kommentarer med forklaringer. Dette vil, sammen med beskrivende variabelnavn, gjøre koden klarere og lettere å forstå. Det vil gjøre det letter i ettertid, og for andre, å forstå hva og hvordan som er tenkt.

```python
P_nominell = 1500    # Kokeplatas oppgitte effekt
U_nominell = 240     # Spenningen som skal gi oppgitt effekt
U_maalt = 226        # Den målte spenningen

# Vi bruker Ohms lov U = R*I og effektlikningen P = U * I
# Så snur vi om på dem for å finne R og sette in for I siden vi ikke kjenner strømmen
# R = U/I og I = P/U => R = U^2/P
R_kokeplate = (U_nominell * U_nominell) / P_nominell  # Beregner kokeplatas resistans.

# Bruker P = U^2/R for å finne effekten ved målt spenning 
P_aktuell_des = (U_maalt * U_maalt) / R_kokeplate  # Beregner aktuell effekt
#  Utregningen gir mange desimaler, så vi bruker funksjonen round()
P_aktuell = round(P_aktuell_des)

print('Kokeplata skal ha en effekt på',P_nominell,'W ved en spenning på',U_nominell,'V')
print('Ved',U_maalt,'V er effekten',P_aktuell, 'W')
```

Vi kan regne ut for andre verdier ved å endre de som tilordnes i koden og kjøre den på nytt. Senere skal vi se hvordan vi får programmet til å be om verdier fra brukeren.

## Tekst, farger og linjetyper

Vi har mange valgmuligheter for å få grafene våre slik vi ønsker. Her er noe av det som kan gjøres:
> Legge til tekst på akser  
> Legge til overskrift  
> Legge til forklaring  
> Endre linjetype eller bare plotte punkter  
> Endre punkttype  
> Endre farge på elementene  
> Legge på rutenett (neste modul)  
> Utheve akser (neste modul). 

### Overskrift og tekst på akser

Første spørsmål fra noen som ser grafene i forige notatbok vil gjerne være: Hva er det vi ser?  
For å slippe slike spørsmål er det greit å legge på overskrift og tekst på aksene.  
For å få gjort dette kaller vi funksjonene ```plt.title()```, ```plt.xlabel()``` og  ```plt.ylabel()```. Teksten som skal stå sendes som argument.

````python
import matplotlib.pyplot as plt
import numpy as np

x_verdier2 = np.linspace(0, 2*np.pi, 36) # 36 verdier fra 0 til 2pi
y_verdier2 = np.sin(x_verdier2)

plt.plot(x_verdier2, y_verdier2)
plt.title("Sinuskurve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
````

### Forklaring til kurvene

Når man har flere kurver er det også lurt å legge på en forklaring til hver kurve slik at man ser hva som er hva. Det gjør man med å legge inn en label i hver plot-kommando og kaller ```plt.legend()```før show.  
Måten vi legger det inn på for hver kurve kalles en nøkkeords argument (Keyword argument = kwarg). Her bruker vi altså nøkkelordet «legend».

````python
x_verdier2 = np.linspace(0, 2*np.pi, 36) # 36 verdier fra 0 til 2pi
sinusverdier = np.sin(x_verdier2)
cosinusverdier = np.cos(x_verdier2)

plt.plot(x_verdier2, sinusverdier, label='sin(x)')
plt.plot(x_verdier2, cosinusverdier, label='cos(x)')
plt.title("Sinus og cosinuskurver")
plt.xlabel("x")
plt.ylabel("sin(x), cos(x)")
plt.legend()
plt.show()
````

### Linjetype, tykkelse og farger

I eksemplene med både sinus- og cosinuskurver ble det automatisk to forskellige farger på kurvene. Men dette kan vi styre selv.  
Vi kan legge inn en formateringsstreng ***fmt*** etter y-verdiene i plot-kommandoen (før eventuell label=). Den består av tre deler: ```markør|linje|farge```.  
Som standard trekkes det linje mellom punktene uten at selve punktene markeres. I noen sammenheng er det lurt å markere punktene. Til og med kan det være aktuelt å ikke ha linje mellom punktene. Markør angis med ett tegn i ***fmt***. Hvis markør utelates blir den ikke tegnet.  
Vi kan velge andre typer linje mellom punktene enn heltrukken. Det er spesielt aktuelt om de skal kunne fungere i svart/hvitt. Linjetypen angis med ett eller to tegn i ***fmt***. Hvis den utelates blir det ikke tegnet noen linje.  
Fargen angis med en bokstav.  

For ovsikt over hvilke tegn/bokstaver som kan brukes og en alternativ måte å gjøre det på, se:  
https://www.w3schools.com/python/matplotlib_markers.asp  

Vi skal prøve det ut i cos-sin plottet vårt. For sinuskurven prøver vi med rød strek-prikk linje uten markører. fmt = '-.r'. For cosinuskurven prøver vi med blå stjerne som markører uten linje. fmt = '*b'

````python
x_verdier2 = np.linspace(0, 2*np.pi, 36) # 36 verdier fra 0 til 2pi
sinusverdier = np.sin(x_verdier2)
cosinusverdier = np.cos(x_verdier2)

plt.plot(x_verdier2, sinusverdier, '-.r', label='sin(x)')
plt.plot(x_verdier2, cosinusverdier, '*b', label='cos(x)')
plt.title("Sinus og cosinuskurver")
plt.xlabel("x")
plt.ylabel("sin(x), cos(x)")
plt.legend()
plt.show()
````

En siste ting vi skal se på her er å endre linjetykkelsen. Det gjør vi med å legge til et nøkkelordsargument der nøkkelordet er *linewith* eller bare *lw*. Verdien bak = er i punktstørrelse.

````python
x_verdier2 = np.linspace(0, 2*np.pi, 36) # 36 verdier fra 0 til 2pi
sinusverdier = np.sin(x_verdier2)
cosinusverdier = np.cos(x_verdier2)

plt.plot(x_verdier2, sinusverdier, lw=2, label='sin(x)')
plt.plot(x_verdier2, cosinusverdier, lw=3, label='cos(x)')
plt.title("Sinus og cosinuskurver")
plt.xlabel("x")
plt.ylabel("sin(x), cos(x)")
plt.legend()
plt.show()
````

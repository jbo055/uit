## Rutenett og akser

De grafene vi så i forrige notatbok var ganske enkel og nakne. Det er mye vi kan legge til og valg vi kan gjøre for å få det bedre.  

Vi skal se å noen nyttige muligheter, men dekker ikke alle.

### Rutenett

Grafene i forrige notatbok kunne det være litt vanskelig å «lese». Det skyldes i stor grad at de mangler akser og andre linjer.  

Det kan vi lett fikse med å legge til rutenet med kommandoen ```plt.grid()```.  
Vi må huske å legge den inn før ```plt.show()```.

````python
import matplotlib.pyplot as plt
import numpy as np

x_verdier1 = np.linspace(0, 2*np.pi, 36) # 36 verdier fra 0 til 2pi
y_verdier1 = np.sin(x_verdier1)

plt.plot(x_verdier1, y_verdier1)
plt.grid()
plt.show()
````

Siden de trigonometriske funksjonene er definert i radiener blir verdiene på x-aksen slik. Uten videre får vi ikke satt x-verdier i brøker av $ \pi $. Det vi kan gøre er lage arrayen med x-verdier i grader og så regne om fra grader til radianer når vi skal gjøre utregningen.

````python
x_verdier2 = np.linspace(0, 360, 36) # 36 verdier fra 0 til 180 grader
y_verdier2 = np.sin(x_verdier2*(np.pi)/180)

plt.plot(x_verdier2, y_verdier2)
plt.grid()
plt.show()
````

Hvis vi nå kan få inndelingene langs x-aksen til å passe bedre blir det bra. Vi bruker kommandoen ```plt.xticks()``` og sender til den en array med de verdiene som skal markeres på x-aksen. Det finnes en tilsvarende yticks for y-aksen.

````python
x_verdier2 = np.linspace(0, 360, 36) # 36 verdier fra 0 til 180 grader
y_verdier2 = np.sin(x_verdier2*(np.pi)/180)

plt.plot(x_verdier2, y_verdier2)
plt.xticks(np.arange(0, 361, 30)) # for å få med markering for 360 må vi ha 361 som parameter
plt.grid()
plt.show()
````

### Utheveing av akser

Det ble lettere med rutenett, men det kan også være greit å få uthevet aksene.  
Dette får vi til med å bruke funksjonene ```axhline()``` og  ```axvline()``` som kan tegne horisontale og vertikale liner i plottet.  

De er altså ikke spesifikt for å tegne akser, men er det som best egner seg for oppgaven.  Funksjonene kan ta flere parameter for å tilpasses slik vi ønsker. Vi skal ikke se på alle, men noen som er aktuelle.  
Først kan vi sette ved vilken verdi (henholdsvis y eller x) linjen skal tegnes. Hvis den utelates er den 0 og faller sammen med aksen slik vi ønsker her.  
De andre parametrene vi skal se på er nøkkelord-argumenter. Det vi sli at vi må ha med _nøkkelord = verdi_ som argument.
> ***color*** eller ***c*** angi farge på linjen farge på linjen.  
> ***linestyle*** eller ***ls*** angi linjestil for eksempel: '-', '--', '-.', ':', ''  
> ***linewith*** eller ***lw*** angi linjetykkelse (float)  

````python
x_verdier2 = np.linspace(0, 360, 36) # 36 verdier fra 0 til 180 grader
y_verdier2 = np.sin(x_verdier2*(np.pi)/180)

plt.plot(x_verdier2, y_verdier2)
plt.xticks(np.arange(0, 361, 30)) # for å få med markering for 360 må vi ha 361 som parameter
plt.grid()
plt.axhline(c='k', lw=2)
plt.axvline(c='k', lw=2)
plt.title("Sinus til vinkel i grader")
plt.xlabel("vinkel (v) i grader")
plt.ylabel("sin(v)")
plt.show()
````

Kurven til tredjegradsligningen blir også bedre med rutenett og uthevede akser.

````python
x_verdier4 = np.arange(-1, 4, .1) # avstand mellom punktene er 0,1. Siste verdi i arrayen blir 3,9
y_verdier4 = 0.5*x_verdier4**3 - 2*x_verdier4**2 + x_verdier4 + 2

plt.plot(x_verdier4, y_verdier4)
plt.grid()
plt.axhline(c='k', lw=2)
plt.axvline(c='k', lw=2)
plt.show()
````

## Plotting, grunnleggende

Tall er vel og bra, men når det blir mange av dem kan det være vanskelig å se for seg sammenenger og variasjoner. Vi ønsker derfor ofte å visualiere dataene. En relativt enkel form for visualisering er å plotte grafer.   

Python har et eget bibliotek (modul) med grunnleggende funkjsoner for å plotte grafer som heter ***Matplotlib***.  
Den funksjonaliteten vi stort sett har bruk for ligger i undermodulen ***pyplot***.  

Vi vil her se hvor elegant og enkelt vi kan plotte grafer.  

For grundigere oversikt og dokumentasjon se:  
https://www.w3schools.com/python/matplotlib_intro.asp  
https://matplotlib.org/stable/

Etter å ha importert pyplot og numpy, som vi også trenger, setter vi opp arrays med x- og y-verdier.  
Vi sender disse til funksjonen ```pyplot.plot()``` og kaller så på funksjonen ```pyplot.show()```.

````python
import matplotlib.pyplot as plt
import numpy as np

x_punkter1 = np.array([0, 5])
y_punkter1 = np.array([0, 250])

plt.plot(x_punkter1, y_punkter1)
plt.show()
````

x_punkter1 og y_punkter1 hadde bare to elementer. Det vil si to punkt. Dermed fikk vi en rett linje.  

Dersom x-aksen bare representerer en nummerering og ikke bestemte verdier er det ikke nødvendig å ha med array for denne. 

````python
y_punkter2 = np.array([3, 0, 2, -1, 4, 1])

plt.plot(y_punkter2)
plt.show()
````

Antall linjer blir en mindre enn antall punkt (verdier i arrayen).

### Plotte funksjoner

Ofte er det grafer av funksjoner som vi vil tegne. Da kan vi regne ut y-verdiene med funksjonen.

````python
x_verdier = np.array([0, 1, 2, 3, 4, 5, 6])
y_verdier = np.sin(x_verdier)

plt.plot(x_verdier, y_verdier)
plt.show()
````

Det ble jo ikke noe bra sinuskurve. Det skyldes at vi har for få punkt langs x-aksen. plot() trekker linjer mellom de punktene vi har. Vi må derfor sørge for at vi har punktene tett nok for å få med de krumninger som grafen gjør.  

Det er derfor lurt å generere arrayen for x-verdiene med ```numpy.arange()``` eller ```numpy.linspace()```. Ved å bruke den arrayen i utregningen av y-verdiene får vi automatisk en array som er like stor. 

````python
x_verdier2 = np.linspace(0, 2*np.pi, 15) # 15 verdier fra 0 til 2pi
y_verdier2 = np.sin(x_verdier2)

plt.plot(x_verdier2, y_verdier2)
plt.show()
````

Det ble bedre, men kurven er fortsatt ikke glatt. til det trengs flere punkt langs x-aksen.  

Et eksempel til som viser samme problemstilling:

````python
x_verdier3 = np.arange(-1, 4, 1) # avstand mellom punktene er 1. Siste verdi i arrayen blir 3
y_verdier3 = 0.5*x_verdier3**3 - 2*x_verdier3**2 + x_verdier3 +2

plt.plot(x_verdier3, y_verdier3)
plt.show()
````

Det skulle være grafen til en tredjegradsligning. Men den ser over hode ikke god ut. Her trengs det mange flere x-verdier.

````python
x_verdier4 = np.arange(-1, 4, .1) # avstand mellom punktene er 0,1. Siste verdi i arrayen blir 3,9
y_verdier4 = 0.5*x_verdier4**3 - 2*x_verdier4**2 + x_verdier4 + 2

plt.plot(x_verdier4, y_verdier4)
plt.show()
````

Det er nesten vanskelig å forestille seg at det er samme likning som er plottet i de to forsøkene. Det viser bare hvor viktig det er å ha nok verdier. Men man skal ikke ha flere en nødvendig heller.

### Flere kurver i samme diagram

Vi kan enkelt plotte flere kurver i samme diagram. Det er da en fordel om de forholder seg til samme område av x-verdier. Vi legger bare til flere kall til ```plt.plot()```.

````python
x_verdier2 = np.linspace(0, 2*np.pi, 36) # 36 verdier fra 0 til 2pi
sinusverdier = np.sin(x_verdier2)
cosinusverdier = np.cos(x_verdier2)

plt.plot(x_verdier2, sinusverdier)
plt.plot(x_verdier2, cosinusverdier)
plt.show()
````

Hvis vi vil plotte flere uavhengige grafer i samme figur kan vi bruke kommandoen ```subplot(r, k, n``` før plottet av hver kurve. Argumentene r og k er henholdsvis antall rader og antall kolonner vi vil ha i figuren. p er i hvilken av dem vi skal plotte.  
Se https://www.w3schools.com/python/matplotlib_subplot.asp for gode eksempel.

## While-løkker

Syntaks for while-løkke:
```Python
while <betingelse>: # Kolon er viktig
    <kommando1>
    <kommando2>
    ...
```

+ Kommandoene som er rykket inn etter ***while*** gjentas så lenge betingelsen er sann
    + Selve betingelsen er som i if-setninger
+ Det som testes i betingelsen (variabelen) må
    + Initialiseres (få en verdi) før løkka
    + Oppdateres i løkka (slik at betingelsen kan bli usann)
+ Utføres 0 eller flere ganger
+ Brukes når
    + man ikke kan bestemme på forhånd antall ganger løkka skal kjøre
    + oppdateringen må skje inne i løkka

#### Flytskjema while-løkke

1. Verdien som skal sjekkes i betingelsen opprettes
2. while – betingelse
3. Dersom betingelsen er sann
4. gjøres programsetningene som er rykket inn  
    inklusive oppdateringen av verdien som inngår i betingelsen
5. Går tilbake til (2) – while
6. Dersom betingelsen er usann
7. hoppes det over alle setningene som er rykket inn. 

Det må programmeres slik at betingelsen til slutt blir usann.  

Eksempler:

````python
teller = 4                # Initialisering av variabel
while teller > 0:         # betingelsen for å kjøre/fortsette i løkka
    print(teller)         # skriver ut tallet som variabelen har nå
    teller = teller - 1;  # oppdatering av tellevariabel
                          # den må nærme seg at betingelsen blir usann
````

Første gang programet kommer til while er betingelsen sann (4 > 0). Derfor kjøres koden som er rykket inn en første gang. I den koden blir teller redusert. Til slutt er teller 0 og d kjøres ikke løkka mer.  


````python
n = int( input('Oppgi tall som skal regnes fakultet av:') )
nStart = n       # Tar vare på n for utskrift

fakultet = 1     # Denne må gis verdi fordi den brukes i løkka
while (n > 0):
    fakultet = fakultet * n  # Bruker her løkke til å regne ut fakultet
                             # n * n-1 * n-2 … * 1
    n = n - 1                # Teller med n
 
print('Fakultet av', nStart, 'er', fakultet)
````

Endrer litt slik at vi kan skrive ut hele utregningen. Bygger opp en resultatstreng ved å bruke f-string.

````python
n = int( input('Oppgi tall som skal regnes fakultet av:') )
resultat = f'Fakultet av {n} er '       # Første del av resultat-strengen

fakultet = 1     # Denne må gis verdi fordi den brukes i løkka
while (n > 0):
    fakultet = fakultet * n  # Bruker her løkke til å regne ut fakultet
                             # n * n-1 * n-2 … * 1
    resultat = resultat + f'{n} * '
    n = n - 1                # Teller med n

resultat = resultat + f'\b\b= {fakultet}' # \b\b fjerner det siste gangetegnet som er lagt til
print(resultat)
````
